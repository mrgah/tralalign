```python
from collatex import *
import string
import re
import graphviz
import csv

def tokenize(input, lg, lgnorms):
    tokens = re.findall('\w+',input)
    print("there are", len(tokens), "tokens of language",lg)
    tokens = [create_token(token, lg, lgnorms) for token in tokens] 
    return tokens

def create_token(input, lg, lgnorms):    
    return {"t":input + " ", "n": normalize(input, lg, lgnorms)}

def read_norms(lg):
    lgnormfile = lg + '.csv'
    norms_out = []

    with open(lgnormfile, 'r') as lgnorms: 
        reader = csv.reader(lgnorms, delimiter=',',quotechar='"')

        for row in reader:
            norm_row = []
            norm_vals = []

            for thing in row:
                if thing.isupper():
                    norm_row.append(thing)
                else:
                    norm_vals.append(thing)
            norm_row[0:0] = [norm_vals]
            norms_out.append(norm_row)
            
                
#     print(lg, "\n", norms_out)
    return norms_out
    
```

```python
def normalize(input, lg, lgnorms):
    input = re.sub('[' + string.punctuation + ']','',input)
        
    for i in range(len(lgnorms)):
        if input in lgnorms[i][0]:
#             print("a match")
#             print(lgnorms[i][1])
            
            tok_al = lgnorms[i][1]
            break
        else:
            tok_al = input.lower()
                
    return tok_al
```

```python
def tralign(lg1, lg2, text):
    
    normalized_tokens = 0
    
    lg1_file = text + '-' + lg1 + '.txt'
    lg2_file = text + '-' + lg2 + '.txt'
     
    with open(lg1_file) as f:
        lg1_text = f.read()
    
    with open(lg2_file) as g:
        lg2_text = g.read()
        
    lg1norms = read_norms(lg1)
    lg2norms = read_norms(lg2)
    
    witness_list = []
    witness_list.append({"id":lg1, "tokens": tokenize(lg1_text, lg1, lg1norms)})
    witness_list.append({"id":lg2, "tokens": tokenize(lg2_text, lg2, lg2norms)})
    
#     print("there are", normalized_tokens,"normalized values")
    
#     print(witness_list)
    
    collation = Collation()
    json_input = {"witnesses": witness_list}
    table = collate(json_input, segmentation=True, layout='vertical',output='html') 
        
```
